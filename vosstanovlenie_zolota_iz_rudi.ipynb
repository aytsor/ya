```python
def train_model(name_model, features, target):
    if name_model == 'dummy':
        param = {}
        model_d = RandomizedSearchCV(DummyRegressor(strategy='mean'), param, n_iter=1, n_jobs=-1,
                                     scoring=scorer_smape, random_state=state, cv=5)
        return model_d.fit(features, target)
    
    if name_model == 'linear':
        param = {}
        model_l = RandomizedSearchCV(LinearRegression(), param, n_iter=1, n_jobs=-1, 
                                     scoring=scorer_smape, random_state=state, cv=5)
        return model_l.fit(features, target)
    
    if name_model == 'tree':
        param = {'max_depth': np.arange(5, 20)}
        model_t = RandomizedSearchCV(DecisionTreeRegressor(random_state=state), param, n_iter=10, n_jobs=-1, 
                                     scoring=scorer_smape, random_state=state, cv=5)
        return model_t.fit(features, target)
    
    if name_model == 'forest':
        param = {'max_depth': np.arange(10, 30, step=2),
                 'n_estimators': np.arange(50, 200, step=10)
                }
        model_f = RandomizedSearchCV(RandomForestRegressor(random_state=state), param, n_iter=30, n_jobs=-1, 
                                     scoring=scorer_smape, random_state=state, cv=3, verbose=2)
        return model_f.fit(features, target)
    
    if name_model == 'boost':
        param = {'learning_rate': np.array([0.01, 0.1, 0.25]), 
                 'n_estimators': np.arange(50, 200, step=10), 
                 'depth': list(np.arange(3, 10)) + [None]
                }
        model_b = RandomizedSearchCV(CatBoostRegressor(random_state=123, verbose=False), param, n_iter=70, n_jobs=-1, 
                                     scoring=scorer_smape, random_state=state, cv=5, verbose=2)
        return model_b.fit(features, target)#–≤—ã–¥–∞–≤–∞–ª–∞—Å—å –æ—à–∏–±–∫–∞ –Ω–∞ random state, –ø–æ—ç—Ç–æ–º—É —è –∑–∞–¥–∞–ª –¥—Ä—É–≥–æ–π. –ö–∞–∫ —ç—Ç–æ —Ñ–∏–∫—Å–∏—Ç—å—Å—è –ø–æ –¥—Ä—É–≥–æ–º—É?
```
